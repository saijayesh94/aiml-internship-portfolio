```python
from PIL import Image
import numpy as np

dinosaur = Image.open('dinosaur.webp')
print(dinosaur.size)
```

```python
numpy_array_dinosaur = np.array(dinosaur)
print(numpy_array_dinosaur)
```

```python
import matplotlib.pyplot as plt

array_dinosaur = plt.imshow(numpy_array_dinosaur)
array_dinosaur
```

```python
gray_scale_image = dinosaur.convert("L")
gray_scale_image
```

```python
resize_image  = dinosaur.resize((100,100))
resize_image.save('resize_dinosaur.jpg')
print(resize_image.size)
```

```python
copy_dinosaur = dinosaur.copy()
resize_image  = copy_dinosaur.thumbnail((100,100))
# resize_image.save('resize_dinosaur_tumbnail.jpg')
copy_dinosaur
```

```python
croped_image = dinosaur.crop((30,40,70,90))
croped_image
```
